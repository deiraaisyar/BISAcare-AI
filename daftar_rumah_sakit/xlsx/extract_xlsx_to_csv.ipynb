```python
import pandas as pd
import os
import glob
from IPython.display import display

folder_path = "data" 

print(f"üìÅ Mengecek folder: {folder_path}")
print(f"Current working directory: {os.getcwd()}")

if os.path.exists(folder_path):
    print(f"‚úÖ Folder ditemukan!")
    files_in_folder = os.listdir(folder_path)
    print(f"üìÑ Isi folder: {files_in_folder}")
else:
    print(f"‚ùå Folder tidak ditemukan: {folder_path}")

excel_files = glob.glob(os.path.join(folder_path, "*.xlsx")) + \
              glob.glob(os.path.join(folder_path, "*.xls"))

print(f"\nüìÅ Ditemukan {len(excel_files)} file Excel:")
for file in excel_files:
    print(f"   - {os.path.basename(file)}")

all_data = []

for file_path in excel_files:
    try:
        print(f"\nüîÑ Membaca: {os.path.basename(file_path)}")
        
        df = pd.read_excel(file_path)
        
        df['Source_File'] = os.path.basename(file_path)
        
        all_data.append(df)
        print(f"   ‚úÖ {df.shape[0]} baris, {df.shape[1]} kolom")
        
    except Exception as e:
        print(f"   ‚ùå Error: {str(e)}")

if all_data:
    combined_df = pd.concat(all_data, ignore_index=True)
    
    output_filename = "daftar_rumah_sakit_allianz.csv"
    combined_df.to_csv(output_filename, index=False, encoding='utf-8')
    
    print(f"\n‚úÖ Berhasil menggabungkan {len(all_data)} file!")
    print(f"üìä Total: {combined_df.shape[0]} baris, {combined_df.shape[1]} kolom")
    print(f"üíæ Tersimpan: {output_filename}")
    
    print(f"\nüìä Preview data gabungan:")
    display(combined_df.head())
    
    print(f"\nüìã Kolom yang tersedia:")
    for i, col in enumerate(combined_df.columns):
        print(f"   {i+1}. {col}")
    
    print(f"\nüìÅ Data per source file:")
    source_counts = combined_df['Source_File'].value_counts()
    for source_file, count in source_counts.items():
        print(f"   - {source_file}: {count} baris")
        
else:
    print("‚ùå Tidak ada data yang berhasil dibaca!")
```

```python
import pandas as pd

df = pd.read_csv('daftar_rumah_sakit_allianz.csv')
df
```

```python
# Ganti nama kolom sesuai yang diminta
df.columns = [
    'Nama_Jaringan',           # 0
    'Penyedia_Layanan',        # 1 (Unnamed: 2)
    'Individu_Kumpulan',       # 2 (Unnamed: 3)
    'Cara_Bayar',              # 3 (Unnamed: 4)
    'Provinsi',                # 4 (Unnamed: 7)
    'Kota',                    # 5 (Unnamed: 8)
    'Telp',                    # 6 (Unnamed: 9)
    'Alamat',                  # 7 (Unnamed: 10)
    'Rawat_Inap',              # 8 (Unnamed: 11)
    'Rawat_Jalan',             # 9 (Unnamed: 12)
    'Rawat_Gigi',              # 10 (Unnamed: 13)
    'Optik'                    # 11 (Unnamed: 14)
]

print("‚úÖ Nama kolom berhasil diganti!")
print(f"üìã Kolom baru: {list(df.columns)}")

# Lihat preview data dengan nama kolom baru
df.head()
```

```python
df['Rawat_Inap'] = df['Rawat_Inap'].astype(str).str.replace('(?i)^ya$', 'Rawat_Inap', regex=True)
df['Rawat_Jalan'] = df['Rawat_Jalan'].astype(str).str.replace('(?i)^ya$', 'Rawat_Jalan', regex=True)
df['Rawat_Gigi'] = df['Rawat_Gigi'].astype(str).str.replace('(?i)^ya$', 'Rawat_Gigi', regex=True)
df['Optik'] = df['Optik'].astype(str).str.replace('(?i)^ya$', 'Optik', regex=True)

df
```

```python
# Rename kolom utama
df = df.rename(columns={'Nama_Jaringan': 'nama_rumah_sakit', 'Alamat': 'alamat', 'Telp': 'telp'})

# Kolom yang akan digabungkan ke Text (kecuali nama_rumah_sakit, alamat, telp)
columns_to_text = [col for col in df.columns if col not in ['nama_rumah_sakit', 'alamat', 'telp']]

print(f"üîÑ Menggabungkan kolom ke Text: {columns_to_text}")

# Fungsi untuk membuat kolom Text yang bersih
def create_clean_text(row):
    """Buat text dari kolom yang dipilih dengan format yang rapi"""
    text_parts = []
    
    for col in columns_to_text:
        value = str(row[col]).strip()
        if value and value != 'nan' and value != 'None' and value != '':
            text_parts.append(value)
    
    return ' | '.join(text_parts)

# Apply fungsi ke setiap baris
df['Text'] = df.apply(create_clean_text, axis=1)

# Bersihkan kolom alamat dan telp dari nilai nan
df['alamat'] = df['alamat'].fillna('').astype(str).str.strip()
df['telp'] = df['telp'].fillna('').astype(str).str.strip()

# Replace 'nan' string dengan empty string
df['alamat'] = df['alamat'].replace('nan', '')
df['telp'] = df['telp'].replace('nan', '')

# Pilih kolom final
result_df = df[['nama_rumah_sakit', 'alamat', 'telp', 'Text']].copy()

# Simpan hasil
result_df.to_csv('daftar_rumah_sakit_allianz_fixed.csv', index=False, encoding='utf-8')
print(f"\nüíæ Data tersimpan: daftar_rumah_sakit_allianz_fixed.csv")

result_df
```

```python
import re

def move_suffix_to_prefix(name):
    """
    Jika nama mengandung koma, pindahkan kata setelah koma ke depan.
    Contoh: 'AWAL BROS BATAM, RS' -> 'RS AWAL BROS BATAM'
    """
    if pd.isna(name) or str(name).strip() == '':
        return name
    name = str(name).strip()
    # Cek pola: ada koma, lalu kata (tanpa koma lagi)
    match = re.match(r'^(.*?),\s*([A-Za-z0-9 ]+)$', name)
    if match:
        main = match.group(1).strip()
        suffix = match.group(2).strip()
        return f"{suffix} {main}"
    return name

# Terapkan ke kolom nama_rumah_sakit
result_df['nama_rumah_sakit'] = result_df['nama_rumah_sakit'].apply(move_suffix_to_prefix)

# Lanjutkan proses seperti sebelumnya
columns_to_text = [col for col in result_df.columns if col not in ['nama_rumah_sakit', 'alamat', 'telp']]

def create_clean_text(row):
    text_parts = []
    for col in columns_to_text:
        value = str(row[col]).strip()
        if value and value != 'nan' and value != 'None' and value != '':
            text_parts.append(value)
    return ' | '.join(text_parts)

result_df['Text'] = result_df.apply(create_clean_text, axis=1)
result_df['alamat'] = result_df['alamat'].fillna('').astype(str).str.strip().replace('nan', '')
result_df['telp'] = result_df['telp'].fillna('').astype(str).str.strip().replace('nan', '')

result_df = result_df[['nama_rumah_sakit', 'alamat', 'telp', 'Text']].copy()
result_df.to_csv('daftar_rumah_sakit_allianz_fixed.csv', index=False, encoding='utf-8')
print("\nüíæ Data tersimpan: daftar_rumah_sakit_allianz_fixed.csv")

result_df
```

```python
import pandas as pd


# Pilih dan rename kolom utama
df = df.rename(columns={
    'Unnamed: 3': 'nama_rumah_sakit',
    'Unnamed: 4': 'alamat',
    'Unnamed: 5': 'telp'
})

# Buat kolom text gabungan
def make_text(row):
    parts = [
        str(row.get('Unnamed: 1', '')).strip(),
        str(row.get('Unnamed: 2', '')).strip(),
        str(row.get('Unnamed: 6', '')).strip(),
        str(row.get('Unnamed: 7', '')).strip(),
        "Manulife"
    ]
    # Gabungkan hanya yang tidak kosong
    return " | ".join([p for p in parts if p and p.lower() != 'nan'])

df['text'] = df.apply(make_text, axis=1)

# Pilih kolom final
result_df = df[['nama_rumah_sakit', 'alamat', 'telp', 'text']].copy()

# Hapus baris yang kosong semua pada kolom utama
result_df = result_df.dropna(subset=['nama_rumah_sakit', 'alamat', 'telp'], how='all')

# Simpan ke CSV
result_df.to_csv("daftar_rumah_sakit_manulife_fixed.csv", index=False, encoding="utf-8")
```

# MSIG

```python
import pandas as pd


df = pd.read_excel("data/msig.xlsx")
```

```python
df = df.drop(df.columns[[0]], axis=1)
df = df.dropna()
df = df[~df.apply(lambda row: row.astype(str).str.contains("CORPORATE", case=False, na=False).any(), axis=1)]
```

```python
# Rename kolom
df = df.rename(columns={
    'Unnamed: 3': 'nama_rumah_sakit',
    'Unnamed: 4': 'alamat',
    'Unnamed: 6': 'telp'
})

# Proses Unnamed:7 (CORPORATE)
df['CORPORATE'] = df['Unnamed: 7'].apply(lambda x: 'CORPORATE' if str(x).strip().upper() == 'YES' else '')

# Proses Unnamed:8 (MCU)
df['MCU'] = df['Unnamed: 8'].apply(lambda x: 'MCU' if str(x).strip().upper() == 'YES' else '')

# Proses Unnamed:9 (INDIVIDU)
df['INDIVIDU'] = df['Unnamed: 9'].apply(lambda x: 'INDIVIDU' if str(x).strip().upper() == 'YES' else '')

# Gabungkan kolom selain nama_rumah_sakit, alamat, telp jadi text + tambah 'MSIG'
exclude_cols = ['nama_rumah_sakit', 'alamat', 'telp']
text_cols = [col for col in df.columns if col not in exclude_cols]

def make_text(row):
    parts = [str(row[col]).strip() for col in text_cols if str(row[col]).strip() and str(row[col]).strip().upper() != 'NAN']
    parts.append('MSIG')
    return ' | '.join(parts)

df['text'] = df.apply(make_text, axis=1)

# Pilih kolom final dan urutkan
result_df = df[['nama_rumah_sakit', 'alamat', 'telp', 'text']].copy()

# Hapus baris yang kosong semua pada kolom utama
result_df = result_df.dropna(subset=['nama_rumah_sakit', 'alamat', 'telp'], how='all')

# Simpan ke CSV
result_df.to_csv("daftar_rumah_sakit_msig_fixed.csv", index=False, encoding="utf-8")
result_df.head()
```

# Prudential

```python
import pandas as pd

df = pd.read_excel("data/prudential_2.xlsx")
```

```python
df = df.drop(df.columns[[0]], axis=1)
df = df.dropna(axis=1, how='all')
```

```python
df = df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 5', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 22'])
df = df.dropna(subset=['Unnamed: 1'])
```

```python
df = df[~df.apply(lambda row: row.astype(str).str.contains("Kategori", case=False, na=False).any(), axis=1)]
```

```python
import re
# Rename kolom utama
df = df.rename(columns={
    'Unnamed: 1': 'nama_rumah_sakit',
    'Unnamed: 8': 'alamat',
    'Unnamed: 17': 'telp'
})

# Gabungkan kolom selain nama_rumah_sakit, alamat, telp menjadi text, hapus "Tanggal bergabung", tambahkan "prudential"
main_cols = ['nama_rumah_sakit', 'alamat', 'telp']
text_cols = [col for col in df.columns if col not in main_cols]

def make_text(row):
    parts = []
    for col in text_cols:
        val = str(row[col]).strip()
        if val and val.lower() != 'nan' and 'Tanggal bergabung' not in val:
            parts.append(val.replace('Tanggal bergabung', '').strip())
    parts.append('prudential')
    return ' | '.join([p for p in parts if p])

df['text'] = df.apply(make_text, axis=1)

# Fungsi untuk memindahkan kata setelah koma ke depan
def move_suffix_to_prefix(name):
    if pd.isna(name) or str(name).strip() == '':
        return name
    name = str(name).strip()
    match = re.match(r'^(.*?),\s*([A-Za-z0-9 ]+)$', name)
    if match:
        main = match.group(1).strip()
        suffix = match.group(2).strip()
        return f"{suffix} {main}"
    return name

df['nama_rumah_sakit'] = df['nama_rumah_sakit'].apply(move_suffix_to_prefix)

# Pilih kolom final dan urutkan
result_df = df[['nama_rumah_sakit', 'alamat', 'telp', 'text']].copy()
result_df
```

```python
result_df = result_df.fillna(" ")
```

```python
result_df.to_csv('daftar_rumah_sakit_prudential_fixed.csv', index=False, encoding='utf-8')
```

# Sompo

```python
import pandas as pd

df = pd.read_excel("data/sompo.xlsx")
```

```python
df = df.drop(df.columns[[0]], axis=1)
```

```python
df = df[~df.apply(lambda row: row.astype(str).str.contains("City", case=False, na=False).any(), axis=1)]
```

```python
# Rename dan urutkan kolom
df = df.rename(columns={
    'Unnamed: 2': 'nama_rumah_sakit',
    'Unnamed: 3': 'alamat',
    'Unnamed: 4': 'telp',
    'Unnamed: 5': 'rawat_inap',
    'Unnamed: 6': 'rawat_jalan'
})

# Proses rawat_inap dan rawat_jalan
df['rawat_inap'] = df['rawat_inap'].apply(lambda x: 'rawat inap' if str(x).strip().upper() == 'Y' else ' ')
df['rawat_jalan'] = df['rawat_jalan'].apply(lambda x: 'rawat jalan' if str(x).strip().upper() == 'Y' else ' ')

# Isi missing value dengan string kosong
df = df.fillna(' ')

# Gabungkan kolom selain nama_rumah_sakit, alamat, telp jadi kolom text, tambahkan kata Sompo
main_cols = ['nama_rumah_sakit', 'alamat', 'telp']
text_cols = [col for col in df.columns if col not in main_cols]

def make_text(row):
    parts = [str(row[col]).strip() for col in text_cols if str(row[col]).strip() and str(row[col]).strip().lower() != 'nan']
    parts.append('Sompo')
    return ' | '.join(parts)

df['text'] = df.apply(make_text, axis=1)
```

```python
result_df = df[['nama_rumah_sakit', 'alamat', 'telp', 'text']].copy()
result_df
```

```python
def clean_telp(telp):
    telp = str(telp).strip()
    if telp and not telp[-1].isdigit():
        return telp[:-1].strip()
    return telp

result_df['telp'] = result_df['telp'].apply(clean_telp)
```

```python
result_df.to_csv('daftar_rumah_sakit_sompo_fixed.csv', index=False, encoding='utf-8')
```

# AIA

```python
import pandas as pd

df1 = pd.read_excel("data/aia_1.xlsx")
df2 = pd.read_excel("data/aia_2.xlsx")
df = pd.concat([df1, df2], ignore_index=True)
df = df.reset_index(drop=True)
```

```python
df = df.drop(index=range(0, 11)).reset_index(drop=True)
```

```python
df = df.drop(df.columns[[0]], axis=1)
```

```python
df = df.dropna(subset=['Unnamed: 3'])
```

```python
df = df.fillna(" ")
```

```python
# Rename dan urutkan kolom
df = df.rename(columns={
    'Unnamed: 3': 'nama_rumah_sakit',
    'Unnamed: 4': 'alamat',
    'Unnamed: 5': 'no_telp',
    'Unnamed: 6': 'rawat_inap',
    'Unnamed: 7': 'rawat_jalan'
})

# Proses rawat_inap: jika Y maka 'rawat inap', selain itu ' '
df['rawat_inap'] = df['rawat_inap'].apply(lambda x: 'rawat inap' if str(x).strip().upper() == 'Y' else ' ')

# Proses rawat_jalan: jika ada Y di Unnamed:7 maka 'rawat jalan', selain itu ' '
df['rawat_jalan'] = df['rawat_jalan'].apply(lambda x: 'rawat jalan' if str(x).strip().upper() == 'Y' else ' ')

# Isi missing value dengan string kosong
df = df.fillna(' ')

# Gabungkan kolom selain nama_rumah_sakit, alamat, no_telp jadi kolom text, tambahkan kata 'aia'
main_cols = ['nama_rumah_sakit', 'alamat', 'no_telp']
text_cols = [col for col in df.columns if col not in main_cols]

def make_text(row):
    parts = [str(row[col]).strip() for col in text_cols if str(row[col]).strip() and str(row[col]).strip().lower() != 'nan']
    parts.append('aia')
    return ' | '.join(parts)

df['text'] = df.apply(make_text, axis=1)

# Pilih kolom final dan urutkan
result_df = df[['nama_rumah_sakit', 'alamat', 'no_telp', 'text']].copy()
result_df
```

```python
result_df = result_df.rename(columns={'no_telp': 'telp'})

def clean_telp(telp):
    telp = str(telp).strip()
    if telp and not telp[-1].isdigit():
        return telp[:-1].strip()
    return telp

result_df['telp'] = result_df['telp'].apply(clean_telp)
```

```python
result_df.to_csv('daftar_rumah_sakit_aia_fixed.csv', index=False, encoding='utf-8')
```

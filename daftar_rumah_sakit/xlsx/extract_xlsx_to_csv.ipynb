```python
import pandas as pd
import os
import glob
from IPython.display import display

folder_path = "data" 

print(f"üìÅ Mengecek folder: {folder_path}")
print(f"Current working directory: {os.getcwd()}")

if os.path.exists(folder_path):
    print(f"‚úÖ Folder ditemukan!")
    files_in_folder = os.listdir(folder_path)
    print(f"üìÑ Isi folder: {files_in_folder}")
else:
    print(f"‚ùå Folder tidak ditemukan: {folder_path}")

excel_files = glob.glob(os.path.join(folder_path, "*.xlsx")) + \
              glob.glob(os.path.join(folder_path, "*.xls"))

print(f"\nüìÅ Ditemukan {len(excel_files)} file Excel:")
for file in excel_files:
    print(f"   - {os.path.basename(file)}")

all_data = []

for file_path in excel_files:
    try:
        print(f"\nüîÑ Membaca: {os.path.basename(file_path)}")
        
        df = pd.read_excel(file_path)
        
        df['Source_File'] = os.path.basename(file_path)
        
        all_data.append(df)
        print(f"   ‚úÖ {df.shape[0]} baris, {df.shape[1]} kolom")
        
    except Exception as e:
        print(f"   ‚ùå Error: {str(e)}")

if all_data:
    combined_df = pd.concat(all_data, ignore_index=True)
    
    output_filename = "daftar_rumah_sakit_allianz.csv"
    combined_df.to_csv(output_filename, index=False, encoding='utf-8')
    
    print(f"\n‚úÖ Berhasil menggabungkan {len(all_data)} file!")
    print(f"üìä Total: {combined_df.shape[0]} baris, {combined_df.shape[1]} kolom")
    print(f"üíæ Tersimpan: {output_filename}")
    
    print(f"\nüìä Preview data gabungan:")
    display(combined_df.head())
    
    print(f"\nüìã Kolom yang tersedia:")
    for i, col in enumerate(combined_df.columns):
        print(f"   {i+1}. {col}")
    
    print(f"\nüìÅ Data per source file:")
    source_counts = combined_df['Source_File'].value_counts()
    for source_file, count in source_counts.items():
        print(f"   - {source_file}: {count} baris")
        
else:
    print("‚ùå Tidak ada data yang berhasil dibaca!")
```

```python
import pandas as pd

df = pd.read_csv('daftar_rumah_sakit_allianz.csv')
df
```

```python
df.drop(columns=['Unnamed: 0', 'Daftar Jaringan untuk Asuransi Individu; Dalam Negeri; Allianz Flexi Medical; Diluar Cakupan Pertanggungan/Perlindungan; Semua Provinsi; Semua Kota;',
                 'Daftar Jaringan untuk Asuransi Individu; Dalam Negeri; Allianz Flexi Medical; Reimbursement; Semua Provinsi; Semua Kota;', 'Source_File', 'Unnamed: 5', 'Unnamed: 6',
                 'Unnamed: 15', 'Unnamed: 16'], inplace=True)
```

```python
df = df[~df.apply(lambda row: row.astype(str).str.contains("Nama Jaringan", case=False, na=False).any(), axis=1)]
df = df.dropna()
```

```python
df.info()
```

```python
# Ganti nama kolom sesuai yang diminta
df.columns = [
    'Nama_Jaringan',           # 0
    'Penyedia_Layanan',        # 1 (Unnamed: 2)
    'Individu_Kumpulan',       # 2 (Unnamed: 3)
    'Cara_Bayar',              # 3 (Unnamed: 4)
    'Provinsi',                # 4 (Unnamed: 7)
    'Kota',                    # 5 (Unnamed: 8)
    'Telp',                    # 6 (Unnamed: 9)
    'Alamat',                  # 7 (Unnamed: 10)
    'Rawat_Inap',              # 8 (Unnamed: 11)
    'Rawat_Jalan',             # 9 (Unnamed: 12)
    'Rawat_Gigi',              # 10 (Unnamed: 13)
    'Optik'                    # 11 (Unnamed: 14)
]

print("‚úÖ Nama kolom berhasil diganti!")
print(f"üìã Kolom baru: {list(df.columns)}")

# Lihat preview data dengan nama kolom baru
df.head()
```

```python
df['Rawat_Inap'] = df['Rawat_Inap'].astype(str).str.replace('(?i)^ya$', 'Rawat_Inap', regex=True)
df['Rawat_Jalan'] = df['Rawat_Jalan'].astype(str).str.replace('(?i)^ya$', 'Rawat_Jalan', regex=True)
df['Rawat_Gigi'] = df['Rawat_Gigi'].astype(str).str.replace('(?i)^ya$', 'Rawat_Gigi', regex=True)
df['Optik'] = df['Optik'].astype(str).str.replace('(?i)^ya$', 'Optik', regex=True)

df
```

```python
# Rename kolom utama
df = df.rename(columns={'Nama_Jaringan': 'nama_rumah_sakit', 'Alamat': 'alamat', 'Telp': 'telp'})

# Kolom yang akan digabungkan ke Text (kecuali nama_rumah_sakit, alamat, telp)
columns_to_text = [col for col in df.columns if col not in ['nama_rumah_sakit', 'alamat', 'telp']]

print(f"üîÑ Menggabungkan kolom ke Text: {columns_to_text}")

# Fungsi untuk membuat kolom Text yang bersih
def create_clean_text(row):
    """Buat text dari kolom yang dipilih dengan format yang rapi"""
    text_parts = []
    
    for col in columns_to_text:
        value = str(row[col]).strip()
        if value and value != 'nan' and value != 'None' and value != '':
            text_parts.append(value)
    
    return ' | '.join(text_parts)

# Apply fungsi ke setiap baris
df['Text'] = df.apply(create_clean_text, axis=1)

# Bersihkan kolom alamat dan telp dari nilai nan
df['alamat'] = df['alamat'].fillna('').astype(str).str.strip()
df['telp'] = df['telp'].fillna('').astype(str).str.strip()

# Replace 'nan' string dengan empty string
df['alamat'] = df['alamat'].replace('nan', '')
df['telp'] = df['telp'].replace('nan', '')

# Pilih kolom final
result_df = df[['nama_rumah_sakit', 'alamat', 'telp', 'Text']].copy()

# Simpan hasil
result_df.to_csv('daftar_rumah_sakit_allianz_fixed.csv', index=False, encoding='utf-8')
print(f"\nüíæ Data tersimpan: daftar_rumah_sakit_allianz_fixed.csv")

result_df
```

```python
import re

def move_suffix_to_prefix(name):
    """
    Jika nama mengandung koma, pindahkan kata setelah koma ke depan.
    Contoh: 'AWAL BROS BATAM, RS' -> 'RS AWAL BROS BATAM'
    """
    if pd.isna(name) or str(name).strip() == '':
        return name
    name = str(name).strip()
    # Cek pola: ada koma, lalu kata (tanpa koma lagi)
    match = re.match(r'^(.*?),\s*([A-Za-z0-9 ]+)$', name)
    if match:
        main = match.group(1).strip()
        suffix = match.group(2).strip()
        return f"{suffix} {main}"
    return name

# Terapkan ke kolom nama_rumah_sakit
result_df['nama_rumah_sakit'] = result_df['nama_rumah_sakit'].apply(move_suffix_to_prefix)

# Lanjutkan proses seperti sebelumnya
columns_to_text = [col for col in result_df.columns if col not in ['nama_rumah_sakit', 'alamat', 'telp']]

def create_clean_text(row):
    text_parts = []
    for col in columns_to_text:
        value = str(row[col]).strip()
        if value and value != 'nan' and value != 'None' and value != '':
            text_parts.append(value)
    return ' | '.join(text_parts)

result_df['Text'] = result_df.apply(create_clean_text, axis=1)
result_df['alamat'] = result_df['alamat'].fillna('').astype(str).str.strip().replace('nan', '')
result_df['telp'] = result_df['telp'].fillna('').astype(str).str.strip().replace('nan', '')

result_df = result_df[['nama_rumah_sakit', 'alamat', 'telp', 'Text']].copy()
result_df.to_csv('daftar_rumah_sakit_allianz_fixed.csv', index=False, encoding='utf-8')
print("\nüíæ Data tersimpan: daftar_rumah_sakit_allianz_fixed.csv")

result_df
```
